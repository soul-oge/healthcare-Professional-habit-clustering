```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
input = pd.read_csv("data_jamlab_isheero_epitech_0125_2.csv", low_memory=False)
```

```python
input.describe()
```

```python
input.head()
```

```python
input.shape
```

```python
for col in ['Segment', 'Generation' ]:
    input[col] = input[col].astype('category')
```

```python
input_dummies = pd.get_dummies(input)
```

```python
input_dummies.head()
```

```python
from sklearn import preprocessing
```

```python
input_preproc = input_dummies[['Digital Affinity','Direct Affinity','Email Affinity','Journal Affinity','Point Of Practice Affinity','Multi-Channel Affinity','Number of Vehicles Reached']]
```

```python
input_preprocd = preprocessing.scale(input_preproc)
```

```python
input_skinny = input_dummies.drop(input_dummies[['Digital Affinity', 'Direct Affinity', 'Email Affinity', 'Journal Affinity', 'Point Of Practice Affinity', 
                               'Multi-Channel Affinity', 'Number of Vehicles Reached']], axis=1)
```

```python
input_skinny.head()
```

```python
input_preprocpd = pd.DataFrame(input_preprocd)
```

```python
input_preprocpd.head()
```

```python
input_preprocpd.columns = ['Digital Affinity', 'Direct Affinity', 'Email Affinity', 'Journal Affinity', 'Point Of Practice Affinity', 
                               'Multi-Channel Affinity', 'Number of Vehicles Reached']
```

```python
input_preprocpd.head()
```

```python
input_skinny.shape
```

```python
input_preprocpd.shape
```

```python
input_analysis = pd.merge(input_skinny, input_preprocpd, right_index=True, left_index=True)
```

```python
input_analysis_final = input_analysis.drop(input_analysis.columns[[0, 1]], axis=1)
```

```python
input_analysis_final.head()
```

```python
input_analysis_final.shape
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(input_analysis_final)

centroid = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroid)
print(labels)
```

```python
labels
```

```python
labels_pred = kmeans.predict(input_analysis_final)
labels_pred
```

```python
from sklearn import metrics
labels_true = labels
metrics.homogeneity_score(labels_true, labels_pred)
```

```python
from sklearn.metrics import adjusted_mutual_info_score
adjusted_mutual_info_score(labels_true, labels_pred)
```

```python
input_clstrd = input
input_clstrd['Cluster Class'] = pd.Series(labels, index = input_clstrd.index)
```

```python
labels_gen = pd.crosstab(index=input_clstrd['Cluster Class'], columns = input_clstrd['Segment'])
labels_gen
```

```python
labels_gen.apply(lambda c: c/c.sum() * 100, axis=1).plot(kind='bar', figsize=(11, 4), stacked =True)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0)
```
